```python
#Part 1-Google Colab and Python
import numpy as np
```

```python
#1. Create a numpy array of size 2 x 2
a=np.array([[1,2],[1,3]])
print(a)
```

```python
#2. Create a random numpy array of size 3 x 4
b=np.random.random((3,4))
print(b)
```

```python
#3. Create a numpy array of ones and a numpy array of zeroes
#zeros
zeros_array=np.zeros((3,3))
print(zeros_array)

#ones
ones_array=np.ones((3,3))
print(ones_array)
```

```python
#4. Create a random numpy array of size 6x3 and transpose it
ori_array=np.random.random((6,3))
transposed_array=ori_array.T
print("Original array:\n")
print(ori_array)
print("\n")
print("Transposed array:\n")
print(transposed_array)
```

```python
#5. Create two random numpy array of size n x p and p x m and then compute the dot product
#get dimension from user
n=int(input("Enter n:"))
p=int(input("Enter p:"))
m=int(input("Enter m:"))

A=np.random.random((n,p))
B=np.random.random((p,m))
C=np.dot(A,B)

print("n x p: \n")
print(A,"\n")

print("p x m: \n")
print(B,"\n")

print("A • B: \n")
print(C)
```
